```python
import os
import numpy as np
import pickle
import sys
import matplotlib.pyplot as plt
sys.path.insert(0, '/Users/angela/src/private/bias-in-datasets/active_learning/src')
from Data import ParsedData
import Parser
from Plotter import write_file, format_plot

```

```python
def plot_losses_hists(experiment_name, experiment_dir, plot_dir):
    
    data_dir = os.path.join(experiment_dir, experiment_name, "pickles/losses")
    if not os.path.isdir(data_dir): return
    
    for filename in os.listdir(data_dir):
        config = Parser.Config(filename, experiment_name)
        print("Writing loss dists for {}".format(filename))
        pickles_file = os.path.join(data_dir, filename)
        with open(pickles_file, 'rb') as handle:
            d = pickle.load(handle)           
            binwidth = 1
            bins = range(int(min(d)), int(max(d)) + binwidth, binwidth)
            n, bins, patches = plt.hist(d,
                                        bins=bins,
                                        normed=0,
                                        facecolor='green',
                                        alpha=0.75)
            
            plt.xlim(0, 20)
            avg = round(np.average(d),2)
            std = round(np.std(d),2)
            title = "mean:{}, std:{}".format(avg, std)
            plt.title(title)
            plot_file_prefix = "{}/losses_{}".format(plot_dir, filename)
            write_file(plot_file_prefix, show=True)
```

```python
plot_home_dir = "../plots"
experiment_dir = "../data/output/cifar10"
experiment_name = "182818_losses"

plot_dir = "{}/{}/".format(plot_home_dir, experiment_name)
if not os.path.exists(plot_dir):
    os.makedirs(plot_dir)
    
plot_losses_hists(experiment_name, experiment_dir, plot_dir)
```
